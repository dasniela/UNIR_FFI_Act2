## SPICE

SPICE es un estándar internacional creado para simular circuitos electrónicos con componentes como: resistencias, diodos, transistores, condensadores, entre otros.

### Historia 
Fue desarrollado en la Universidad de Berkley ya que se necesitaba un simulador de circuitos. 

Previamente se había creado un simulador de circuitos llamado CANCER en la decada de los 60's, posteriormente fue reescrito y renombrado como SPICE bajo el liderazgo de el profesor Donald Pderson, el cual decidió lanzarlo ahora como un programa de dominio público ya que pensaba que todo el proceso significativo ocurre cuando la información es compartida libremente.

El análisis .TRAN (transient) es el que hace un anállisis transitorio desde tstart (que por defecto es 0) hasta tstop 

La sintaxis es:

.TRAN TSTEP=<float> TSTOP=<float> [TSTART=<float>  UIC=0/1/2/3 [IC_LABEL=<string>] METHOD=<string>]

UIC es un parámetro que nos sirve para especificar el estado del circuito en un tiempo t=tstart y puede tener 4 valores:

|UIC| DESCRIPCIÓN |
|:--:|:--:|
|uic=0| todos los voltajes y corrientes de los nodos a través de las fuentes se asumen a cero en tstart 
|uic=1| el estatus de t'=strat el es último resultado del análisis OP 
|uic=2|  el estado en t = tstart es el último resultado de un análisis OP en el que se establecen los valores de las corrientes a través de inductores y voltajes en los condensadores especificados en su ic (ic son las condiciones iniciales para el análisis de un circuito) |
|uic=3|  Cargar un ic proporcionado por el usuario. Esto requiere una directiva .ic en algún lugar de la lista de redes |

## Comprobación de la ecuación:

Para no calcular resultados basados en una mentira enorme comprobaremos la siguiente ecuación básica de este tema:
$$
V = I\cdot R
$$
Para ello averiguaremos primero las dimensiones de cada magnitud, trabajando con sympy:


```python
from sympy.physics.units import voltage
from sympy.physics.units import volt
from sympy.physics.units import current
from sympy.physics.units import ampere
from sympy.physics.units import impedance
from sympy.physics.units import ohm
from sympy.physics.units.systems.si import dimsys_SI
```

A continuación sacaremos las dimensiones del voltaje:

```python
dimsys_SI.get_dimensional_dependencies(voltage)
```

Significa que la dimensión formada por las dimensiones de la corriente (I) por las dimensiones de la resistencia (R) debe de estár compuesta por las siguientes:
$$
\mathrm{\Big[V\Big] = \bigg[\frac{M\cdot L^2}{I\cdot T^3}\bigg]}
$$
Ahora averiguamos las dimensiones de la corriente y la resistencia:

```python
dimsys_SI.get_dimensional_dependencies(current)
```

```python
dimsys_SI.get_dimensional_dependencies(impedance)
```

Como vemos I y R tienen las siguientes dimensiones:
$$
\require{cancel}\mathrm{\Big[I\Big] = \Big[I\Big]\qquad \qquad \Big[R\Big] = \bigg[\frac{M\cdot L^2}{I^2\cdot T^3}\bigg]}
$$

Multiplicando ahora las dos dimensiones nos sale la misma dimensión que tiene el voltaje:

$$
\mathrm{\Big[R\cdot I\Big] = \bigg[ \cancel{I} \cdot \frac{M\cdot L^2}{I^\cancel{2}\cdot T^3}\bigg] = \bigg[\frac{M\cdot L^2}{I\cdot T^3}\bigg] = \Big[V\Big]}
$$

#### Ley de Ohm     

La cantidad de corriente es proporcional al voltaje e inversamente proporcional a la resistencia

$$ I = \frac{V}{R} $$

Más arriba ya comprobamos que esa expresión es correcta en cuanto a las dimensiones.

Para empezar con los cálculos importamos lo necesario de Sympy:



```python
from sympy import solve, symbols, Eq
from sympy. physics.units import convert_to
```

Cogemos como valores:
$V = 9\,\mathrm{V}$ y $R = 100\,\mathrm{\Omega}$

Calculamos la intensidad para esos parámetros:

```python
R = 100 * ohm
V = 9 * volt
I = symbols("i")
eq_ohm = Eq(I, V/R)
solucion_I = solve(eq_ohm, I)[0]
I = solucion_I
convert_to(I, [ampere]).n(2)
```

Por lo tanto los resultado corresponden con el análisis `op`

> **Pregunta**: Demuestra que se cumple la Ley de Kirchhoff de la energía en un circuito, es decir, que la suma de la energía suministrada por las fuentes (pilas) es igual a la consumida por las resistencias. Realiza la operación con Sympy.

$$
\sum_i^N V_{\text{fuentes}} = \sum_j^M V_{\text{consumido en resistencias}}
$$

Osea:

$$
V = V_1 + V_2 + V_3 = 9 \mathrm{\,V}
$$

Lo comprobamos con Sympy:

```python
V_total = 9.0 * volt
I = 0.0005 * ampere
R1 = 3*kilo * ohm
R2 = 10*kilo * ohm
R3 = 5*kilo * ohm
V = symbols('V_consumido')
V1 = symbols('V_1')
V2 = symbols('V_2')
V3 = symbols('V_3')
ley_Kirchhoff = Eq(V, V1+V2+V3)
ley_ohm_V1 = Eq(V1, I * R1)
ley_ohm_V2 = Eq(V2, I * R2)
ley_ohm_V3 = Eq(V3, I * R3)
solucion_V = solve([ley_ohm_V1, ley_ohm_V2, ley_ohm_V3, ley_Kirchhoff], [V1, V2, V3, V])
V_consumida = solucion_V
V_consumida = convert_to(V_consumida, [volt])
if V_total == V_consumida["V_consumido"]:#comprobamos si el potencial calculado por Ahkab es igual al calculado por nosotros.
    print("La energía suministrada es igual a la consumida: ", V_consumida["V_consumido"])
```

## Comprobación de los resultados de la simulación:

Mediente la Ley de Kirchhoff comprobaremos los resultados obtenidos por la simulación.
$$
I_{\text{entrante}} = \sum_i^{N} I_{\text{salientes}} \quad \text{y} \quad \sum_i^N V_{\text{fuentes}} = \sum_j^M V_{\text{consumido en resistencias}}
$$

Para ello utilizaremos Sympy:

```python
# poner todo en valor absolut
intensidad_total = -(resultados['op']['I(VDD)'][0][0]*ampere).n(3)
intensidad_1 = resultados['op']['I(VDUMMY1)'][0][0]*ampere
intensidad_2 = resultados['op']['I(VDUMMY2)'][0][0]*ampere
I = symbols('Intensidad_total')
ley_Kirchhoff_corriente = Eq(I, intensidad_1 + intensidad_2)
intensidad_total_ley = solve(ley_Kirchhoff_corriente, I)[0].n(3)
if intensidad_total == intensidad_total_ley: #comprobamos si la intensidad calculada por Ahkab es igual a la calculada por nosotros.
    print("Vemos que la corriente total coincide con la suma de las corrientes individuales. Se cumple la ley de Kirchhoff para corrientes. I_total = ", intensidad_total_ley)
```

Ahora comprobada la corriente podemos seguir con la comprobación de los voltajes devolvidos por Ahkab:

```python
# poner todo en valor absoluto
v_total = -(resultados['op']['V1'][0][0]*volt)
I_total = -resultados['op']['I(VDD)'][0][0]*ampere
I2 = resultados['op']['I(VDUMMY1)'][0][0]*ampere
I1 = resultados['op']['I(VDUMMY2)'][0][0]*ampere
r2 = 1*kilo*ohm
r3 = 220*ohm
r4 = 1.5*kilo*ohm
r5 = 470*ohm
v2 = symbols('V2')
v3 = symbols('V3')
v4 = symbols('V4')
v5 = symbols('V5')
ley_ohm_v2 = Eq(v2, I_total*r2)
v2_consumida = solve(ley_ohm_v2, v2)[0]
v2_consumida = convert_to(v2_consumida, [volt])
#v2 representa potencial restante, osea, potencial que queda despues de que r2 haya consumido lo necesario
v2 = (v_total-v2_consumida).n(6)
print("V2 = ",v2)
ley_ohm_v3 = Eq(v3, I1*r3)
v3_consumida = solve(ley_ohm_v3, v3)[0]
v3_consumida = convert_to(v3_consumida, [volt])
#v3 representa potencial restante, osea, potencial que queda despues de que r3 haya consumido lo necesario
v3 = (v2-v3_consumida).n(5)
print("V3 = ",v3)
ley_ohm_v4 = Eq(v4, I1*r4)
v4_consumida = solve(ley_ohm_v4, v4)[0]
v4_consumida = convert_to(v4_consumida, [volt]).n(5)
#v4 representa potencial consumido por r4
print("V4 = ", v4_consumida)
ley_ohm_v5 = Eq(v5, I2*r5)
v5_consumida = solve(ley_ohm_v5, v5)[0]
v5_consumida = convert_to(v5_consumida, [volt]).n(6)
#v5 representa potencial consumido por r5
print("V5 = ", v5_consumida)
```

Se ve que los cálculos de Ahkab coinciden con los nuestros.

Comprobaremos además, que la Ley de Kirchhoff para los voltaje se cumpla.

Para que se cumpla, la suma del potencial consumido por cada elemento en el circuito, debe ser iigual al potencial suministrado por la fuente (pila):

```python
#en este caso nos hace falta redondear el resultado, porque no da 0 exactamente
import math
ley_Kirchhoff_1 = math.trunc((v_total-v2_consumida-v3_consumida-v4_consumida)/volt)
ley_Kirchhoff_2 = math.trunc((v_total-v2_consumida-v5_consumida)/volt)
print("Voltaje rama 1:", ley_Kirchhoff_1, "\nVoltaje rama 2:", ley_Kirchhoff_2)
```

Como podemos ver todos dan 0, por lo que se comprueba la Ley de Kirchhoff para el voltaje.

En cambio en la gráfica del voltaje podemos ver que el voltaje se incrementa desde cero Voltios hasta el valor de la fuente de corriente conforme el condensador se carga de igual manera, al llegar a la constante de tiempo el condensador, esta totalmente llena la capacidad del condensador, aunque se sigue cargando mientrás el circuito este conectado y por eso vemos que despues de 5 s la curva se estabiliza. Esto lo podemos confirmar calculando el tiempo de carga del condensador:

$𝑇=5\cdot 𝑅\cdot 𝐶$
 
T: Consante de tiempo en la cual se logra el 99.3% de la carga total

R: Resistencia ( $\mathrm{\Omega}$ )

C: Capacidad del condensador ( $\mathrm{𝐹}$ )

$𝑇=\mathrm{5⋅1000\Omega\cdot 0.001𝐹=5𝑠}$
 
¿Por qué multiplicamos la constante de tiempo por 5?
En cada constante de tiempo iniciando en 1 y acabando en 5 los porcentajes de voltahe incrementan y los de intensidad disminuyen y al mismo tiempo podemos ver que la suma del voltaje y la corriente en cada momento de la constante de tiempo (1-5) suman el 100%

CONSTANTE TIEMPO | PORCENTAJE DE VOLTAJE |PORCENTAJE DE CORRIENTE|
:--:|:--:|:--|
1|63.2%|36.8%|
2|86.5%|13.5%|
3|95%| 5%|
4|98.2%| 1.8%|
5| 99.3%| 0.7%|


Es decir, que cuando el tiempo llega a 5 segundos el condensador esta al 99.3% de la carga y es cuando podemos ver en la gráfica que la corriente baja a casi cero.

El tiempo de carga como vemos, no depende del voltaje; no importa si el voltaje es mayor o menor porque el tiempo de carga depende de la resistencia y del condensador

>¿Por qué son opuestas?

Son opuestas porque en t=0 la intensidad es la máxima y va disminuyendo y el Voltaje inicia en cero y va aumentando .

>¿Qué valores devuelve el análisis de tipo `.op`?

El análisis .op devuelve el voltaje e intensidad de cada rama. En este caso la intensidad que nos devuelve es la de V1.
